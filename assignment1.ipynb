# NYC Taxi Demand Forecasting

In this notebook, I explore forecasting hourly taxi trip demand in New York City using models covered in Weeks 1â€“3 of Econ 8310.

The goal is to build a forecasting model that:
- Uses historical hourly taxi data
- Produces forecasts for one future month (744 hours)
- Performs well based on RMSE
- Can be exported into a `.py` script for final submission


```python
import pandas as pd
import numpy as np

from statsmodels.tsa.holtwinters import ExponentialSmoothing

train_url = "https://github.com/dustywhite7/econ8310-assignment1/raw/main/assignment_data_train.csv"
train = pd.read_csv(train_url)

train.head()
```

```python
y = train["trips"].astype(float)

model = ExponentialSmoothing(
    y,
    trend="add",
    seasonal="mul",
    seasonal_periods=24
)
```

```python
modelFit = model.fit(optimized=True)
```

```python
pred = modelFit.forecast(744)
pred = np.array(pred, dtype=float)

len(pred)
```
